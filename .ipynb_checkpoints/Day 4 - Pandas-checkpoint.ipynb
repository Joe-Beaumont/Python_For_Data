```python
#Cleaning the Data

df_messy = pd.read_csv("/Day 4/students_data_messy.csv")
df_messy.head()
```

```python
# drop rows where any value is missing
df_dropped = df_messy.dropna()
df_dropped.head()

#drop rows where a specific column is missing
df_dropped_col = df_messy.dropna(subset=["Age"])
df_dropped_col.head()
```

```python
# fill the missing values with a constant
df_messy["Age_Filled"] = df_messy["Age"].fillna(0)
df_messy.head()

#fill missing values with a mean or median or statistic
df_messy["Score_Filled"] = df_messy["Score"].fillna(df_messy["Score"].mean())
df_messy.head()
```

```python
#Challenge 6
#Filtering & Sorting
#Filter multiple conditions use and or, age and score, attendance and extra curricular
print(df_messy[(df_messy["Age"] > 17) & (df_messy["Attendance"] > 90)])
```

```python
#Sorting

#Sort by Score ascending
score_ascending = df_messy.sort_values(by="Score")
#Sort by Score Descending
score_descending = df_messy.sort_values(by="Score", ascending = False)
#Sort by age (ascending) and score (descending) - Not quite right
age_and_score = df_messy.sort_values(by=["Score", "Age"], ascending = [False, True])
```

```python
#renaming columnbs

#rename specific columns
df_renamed = df_messy.rename(columns={"Extracurricular" : "Club"})
df_renamed.head()

#rename all the columns in one go

df_messy.head()
df_messy.columns = ["Name", "School_Name", "Grade_Level", "Test_Score", "Age_Years", "Attendance_Rate", "Club", "Email_Address", "Age_Extra", "Score_Mean"]
df_messy.head()
```

```python
#Challenge 6:
df_cars = pd.read_csv("/Day 4/messy_cars.csv")
df_cars.head()
#Drop or fill missing values appropriately.
df_cars["Money_Filled"] = df_cars["Money"].fillna(df_cars["Money"].mean())
df_cars.head()
#Filter the dataset for a meaningful subset.
print(df_cars[df_cars["Money"] >30000])
#Sort the data by miles descending.
df_cars_sorted = df_cars.sort_values(by="Miles", ascending = False)
df_cars_sorted.head()
#Rename columns for clarity.
df_cars_sorted.columns = ["License_Plate", "Make", "Model", "Year_Manufactured", "Cost", "Mileage", "Colour", "Cost_Filled"]
df_cars_sorted.head()
#Save the cleaned dataset as cars_cleaned.csv.
df_cars_sorted.to_csv("/home/joe-beaumont/PyCharmMiscProject/cars_cleaned.csv", index = False)
```

```python
#GROUP BY & Multiple Aggregations
```

```python
#Group by brand and select one column before the aggregation
brand_group = df_cars.groupby("Brand")["Miles"]
brand_group.head()
```

```python
brand_group = df_cars.groupby("Brand")["Miles"].mean().reset_index(name = "Average_Mileage")
brand_group.head()
```

```python
# Multiple Aggregations on the brand: money, car_num, 
brand_stats = df_cars.groupby("Brand").agg(
    avg_price=("Money", "mean"),
    avg_mileage=("Miles", "median"),
    total_cars=("Car_num", "count")
).reset_index()
brand_stats.head()
```
