Seaborn

```python
import seaborn as sns
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
tips["tip %"] = tips["tip"]/tips["total_bill"] * 100
```

```python
sns.scatterplot(x = 'tip', y = 'total_bill', data = tips)
```

```python
sns.scatterplot(x = 'day', y = 'tip', data = tips, hue = 'sex', palette='YlGnBu')
```

```python
sns.barplot(x = 'day', y ='tip %', data = tips, hue = 'day', palette='YlGnBu')
```

```python
sns.histplot(tips['tip'], kde=True, bins=5)
```

```python
sns.barplot(x='sex', y='tip', data = tips, hue='sex', palette='YlGnBu')
```

```python
sns.boxplot(x = 'day', y = 'tip', data = tips, hue = 'sex', palette='YlGnBu')
```

```python
sns.stripplot(x = 'day', y = 'tip', data = tips, hue = 'sex', palette='YlGnBu')
```

```python
sns.jointplot(x='tip', y='total_bill', data = tips, kind = 'reg')
```

```python
sns.pairplot(data = tips, x_vars = ["sex", "day", "smoker", "time", "size"], y_vars = ['tip', 'tip %'], hue="day", diag_kind="hist")
sns.pairplot(data=tips, x_vars=["sex", "day", "smoker", "time", "size"], y_vars=["tip %", "tip"], hue="day", diag_kind="hist")
```
