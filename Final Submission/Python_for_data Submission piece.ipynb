```python
import seaborn as sns
import pandas as pd
sns.color_palette("mako", as_cmap=True)
```

```python
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
total_survivors = titanic["survived"].sum()
print(total_survivors)
```

```python
sns.barplot(x = 'sex', y = 'survived', hue = 'sex', palette = 'mako', data = titanic)
```

As we can see the survival rate was much higher for females

```python
sns.barplot(x = 'class', y = 'fare', data = titanic, hue = 'class', palette = 'mako')
```

As would be expected first class tickets were significantly more expensive

```python
sns.barplot(x = 'class', y = 'survived', data = titanic, hue ='sex', palette = 'mako')
```

Those in first class had a much higher rate of survival

```python
sns.stripplot(x = 'survived', y = 'age', data = titanic, hue = 'age', palette = 'mako')
```

There doesn't seem to be a link between age and survival

```python
sex_count = titanic.groupby('sex').size()
print(sex_count)
```

```python
class_count = titanic.groupby('class').size()
print(class_count)
```

```python
sns.stripplot(x='class', y='age', data = titanic, hue = 'survived', palette = 'mako')
```
